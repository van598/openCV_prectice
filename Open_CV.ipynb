<h1> Image Rotation and flipping </h1>

```python
if image is None:
    print("Error: Image not loaded.")
else:
    h, w, c = image.shape
    print(f"Height: {h}, Width: {w}, Channels: {c}")
    center_x, center_y = w // 2, h // 2
    F=cv2.getRotationMatrix2D((center_x, center_y), 45, 1.0)
    rotated_image=cv2.warpAffine(image, F, (w, h))
    cv2.imshow("Rotated Image", rotated_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if image is None:
    print("Error: Image not loaded.")
else:
    flip=cv2.flip(image, 1) #horizontal flip
    cv2.imshow("Flipped Image", flip)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
```

```python
cap_height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
cap_width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
codec=cv2.VideoWriter_fourcc(*'XVID')
recoder=cv2.VideoWriter('my_video.avi', codec, 20, (cap_width, cap_height))
bg_subtractor=cv2.createBackgroundSubtractorMOG2()
while True:
    sucess, frame=cap.read()
    if not sucess:
        break
    recoder.write(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))
    fg_mask=bg_subtractor.apply(frame)
    #remove noise and fill in the holes using dilation
    kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
    fg_mask=cv2.erode(fg_mask, kernel, iterations=1)
    fg_mask=cv2.dilate(fg_mask, kernel, iterations=1)
    #find contours of the moving objects
    capture_contours, heirarchy = cv2.findContours(fg_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    #capture bigger contours
    for contour in capture_contours:
        if cv2.contourArea(contour) <1000:
            continue
        #draw bounding ractangles around the moving objects
        (x, y, w, h) = cv2.boundingRect(contour)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.imshow("Video Recording", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break    
cap.release()
recoder.release() 
cv2.destroyAllWindows()   
```
